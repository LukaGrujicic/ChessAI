```python
class Node:
    #N: Number of times parent node is visited
    #n: Amount of time current node is visited
    #v: Value of node 
    def __init__(self, board=None, N=None):
        self.board = board
        self.v = 0
        self.n = 0
        self.N = N
        self.children = []

    def calc_children(self):
        if not self.board.is_game_over():
            for move in self.board.legal_moves:
                p_board = chess.Board(self.board.fen())        
                p_board.push(move)
                currChild = Node(p_board,self.n)
                self.children.append(currChild)             
     
    def calc_ucb(self,N):
        C = 0.01
        rt = (self.v/self.n)+C*(np.sqrt(np.log(N)/self.n))
        return rt
```
